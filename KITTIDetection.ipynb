```python
from ultralytics import YOLO
import warnings

# Suppress all the deprecation warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)

model = YOLO("KITTI.pt")

# --- 1. Run Validation for IoU 50 ---
print("--- Running validation for IoU 50 ---")
# The variable metrics_iou50 is defined here.
metrics_iou50 = model.val(
    data="./dataKITTI.yaml",
    imgsz=640,
    batch=4,
    workers=0,
    device=0,
    cache=True,
    amp=True,
    project='runs', name='KITTIDET_IOU50',
    iou=0.50
)

# --- 2. Run Validation for IoU 70 ---
print("\n--- Running validation for IoU 70 ---")
# The variable metrics_iou70 is defined here.
metrics_iou70 = model.val(
    data="./dataKITTI.yaml",
    imgsz=640,
    batch=4,
    workers=0,
    device=0,
    cache=True,
    amp=True,
    project='runs', name='KITTIDET_IOU70',
    iou=0.70
)

# --- 3. Calculate Class Indices (Must be run before using *_index) ---
class_names = metrics_iou50.names
names_list = list(class_names.values())

try:
    # The variables car_index and pedestrian_index are defined here.
    car_index = names_list.index('Car')
    pedestrian_index = names_list.index('Pedestrian')

    print(f"\nFound indices: Car: {car_index}, Pedestrian: {pedestrian_index}")
    
except ValueError as e:
    print(f"\nError: Could not find one of the required classes ('Car' or 'Pedestrian'). Details: {e}")
    # Handle the error gracefully if the classes aren't found
    # You might need to exit or set indices to -1 here.
```
